```python
from sklearn.cluster import KMeans
import pandas as pd

df=pd.read_csv('cardata.csv')
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Assuming df is already the DataFrame with your data
plt.figure(figsize=(8,6))

# Plotting the scatter plot for Selling_Price vs Present_Price
plt.scatter(df['Selling_Price'], df['Present_Price'], color='purple')

# Adding labels and title
plt.xlabel('Selling Price (in Lakhs)')
plt.ylabel('Present Price (in Lakhs)')
plt.title('Selling Price vs Present Price')

# Show plot
plt.show()

```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


x = df[['Selling_Price', 'Present_Price']].values


inertias = []


for i in range(1, min(6, len(x)+1)): 
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(x)  
    inertias.append(kmeans.inertia_)  

# Plotting the Elbow method graph
plt.plot(range(1, len(inertias)+1), inertias, marker='o')
plt.title('Elbow Method For Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(df[['Selling_Price', 'Present_Price']])

plt.scatter(df['Selling_Price'], df['Present_Price'], c=kmeans.labels_)
plt.show()
```
