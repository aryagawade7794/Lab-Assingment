```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('cardata.csv')
```

```python
print(df)
```

```python
x= df.iloc[ : ,1:5 ].values
y= df.iloc[ : , 3 ].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 1/3 , random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

# Feature Scaling for x_train and x_test
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)

# Reshaping y_train to make it a 2D array
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train.reshape(-1, 1))  # Reshape y_train to 2D
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Predicting the Test set results
y_pred = regressor.predict(x_test)
```

```python
import matplotlib.pyplot as plt

# Visualising the Training set results
# If x_train is 2D, make it 1D for plotting
plt.scatter(x_train[:, 0], y_train, color='green')  # Use x_train[:, 0] if x_train has more than one feature
plt.plot(x_train[:, 0], regressor.predict(x_train), color='blue')

# Visualising the Test set results
# Again, ensure x_test is 1D
plt.scatter(x_test[:, 0], y_test, color='red')  # Use x_test[:, 0] for 2D input
plt.plot(x_train[:, 0], regressor.predict(x_train), color='blue')

plt.title('Present Price vs Parameter (Test Set)')
plt.xlabel('Present Price')
plt.ylabel('Parameter')
plt.show()
```
